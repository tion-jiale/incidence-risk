# **FEATURE SELECTION**

BINNING THE TARGET VARIBALE

```python
quartiles = pd.qcut(final_data['incidence'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

def map_quartiles_to_custom_categories(q_label):
    if q_label == 'Q1':
        return 'Low'
    elif q_label in ['Q2', 'Q3']:
        return 'Moderate'
    elif q_label == 'Q4':
        return 'High'
    return None

final_data['incidence_quartile_custom'] = quartiles.apply(map_quartiles_to_custom_categories)

print("Value counts for 'incidence_quartile_custom':")
print(final_data['incidence_quartile_custom'].value_counts())

final_data
```

```python
encoding_map = {
    'Low': 0,
    'Moderate': 1,
    'High': 2
}
final_data['incidence_encoded'] = final_data['incidence_quartile_custom'].map(encoding_map)
final_data
```

```python
final_data.info()
```

```python
#Assign target variable
X = final_data.drop(columns=['incidence_encoded','incidence_quartile_custom'])
y = final_data['incidence_encoded']
```

```python
#Split into numerical and categorical features
numerical_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['bool', 'object']).columns
```

```python
from sklearn.feature_selection import f_classif

X_numerical = X[numerical_features]
anova_results = []
for feature in X_numerical.columns:
  F, p = f_classif(X_numerical[[feature]], y)
  anova_results.append({'Feature': feature, 'F-value': F[0], 'p-value': p[0]})
anova_results = pd.DataFrame(anova_results).sort_values(by='p-value')
print("\nANOVA Results (Numerical Features):")
display(anova_results)
```

```python
from scipy.stats import chi2_contingency

# Update y to use the latest custom binned target
y = final_data['incidence_encoded']



# Identify categorical features for Chi-squared test (one-hot encoded boolean columns)
categorical_features_for_chi2 = X.select_dtypes(include=['bool']).columns.tolist()

chi2_results = []
for feature in categorical_features_for_chi2:
    contingency_table = pd.crosstab(X[feature], y)
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    chi2_results.append({'Feature': feature, 'Chi2 Statistic': chi2, 'p-value': p})

chi2_results_df = pd.DataFrame(chi2_results).sort_values(by='p-value')

print("\nChi-squared Test Results (Categorical Features):")
display(chi2_results_df)
```

```python
# Select top features based on threshold

selected_numerical = anova_results[anova_results['p-value'] < 0.05]['Feature'].tolist()
selected_categorical = chi2_results_df[chi2_results_df['p-value'] < 0.05]['Feature'].tolist()
```

```python
#Combine Selected Features
selected_features = selected_numerical + selected_categorical
print("\nSelected Features:")
print(selected_features)

#Create final dataset

X_dataset = X[selected_features]
```

# ROC and AUC

# **Holdout method**

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import roc_curve, auc, roc_auc_score
from sklearn.preprocessing import LabelBinarizer

# Binarize y_test for multiclass ROC plotting
lb = LabelBinarizer()
y_test_binarized = lb.fit_transform(y_test)
n_classes = y_test_binarized.shape[1]

# Get predicted probabilities for each model
y_score_log = lr.predict_proba(X_test_scale)
y_score_rf = rf_model_updated.predict_proba(X_test_scale)
y_score_xgb = xgb_model.predict_proba(X_test_scale)
y_score_cb = cb_model.predict_proba(X_test_scale)

# Dictionary of model predicted probabilities
# Ensure all prob_ variables are defined and accessible from previous steps
auc_models = {
    "Logistic Regression": y_score_log,
    "Random Forest": y_score_rf,
    "xG Boost": y_score_xgb,
    "CatBoost": y_score_cb,
}

auc_results = {}
```

```python
plt.figure(figsize=(12, 8))
plt.plot([0, 1], [0, 1], 'k--', label='Chance (AUC = 0.50)')

print("=== AUC Scores and ROC Curves for All Models ===\n")

for model_name, prob_pred in auc_models.items():
    # Calculate macro-average AUC score for reporting
    macro_auc = roc_auc_score(y_test_binarized, prob_pred, multi_class='ovr', average='macro')
    auc_results[model_name] = macro_auc

    # For plotting a single representative ROC curve for multiclass, use micro-average approach
    # Flatten the binarized true labels and predicted probabilities for micro-average
    fpr, tpr, _ = roc_curve(y_test_binarized.ravel(), prob_pred.ravel())
    micro_auc = auc(fpr, tpr) # This auc will be for the micro-average curve

    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {macro_auc:.2f})')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
# Convert to DataFrame for clean display
df_auc = pd.DataFrame.from_dict(auc_results, orient="index", columns=["AUC"])
df_auc = df_auc.sort_values(by="AUC", ascending=False)

print("\n=== Consolidated AUC Scores ===")
print(df_auc)
```
